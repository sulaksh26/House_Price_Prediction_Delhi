```python
import pandas as pd
import numpy as np
```

```python
data.drop(columns=['Furnishing','Parking','Status','Transaction','Type'],inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop(columns=['Per_Sqft'],inplace=True)
```

```python
data['price_per_sqft']=data['Price']/data['Area']
```

```python
data.head()
```

```python
data['Locality'].value_counts()
```

```python
data['Locality']=data['Locality'].apply(lambda x:x.strip())
locality_count=data['Locality'].value_counts()
```

```python
locality_count_less_than_2=locality_count[locality_count<2]
locality_count_less_than_2
```

```python
data['Locality']=data['Locality'].apply(lambda x: 'other' if x in locality_count_less_than_2 else x)
```

```python
data.head()
```

```python
data['Locality'].value_counts()
```

```python
data.drop(columns=['price_per_sqft'],inplace=True)
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data=data.dropna(subset=['Bathroom'])
```

```python
data.isna().sum()
```


**Cleaned Data**

```python
data.head()
```

```python
data.to_csv('Cleaned-data.csv')
```

```python
y=data['Price']
X=data.drop(columns=['Price'])
```

**Applying ML Techniques**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

**Applying Linear Regression**

```python
column_trans=make_column_transformer((OneHotEncoder(sparse=False),['Locality']),remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

**Applying Lasso**

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lasso)
```

**Applying Ridge**

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_ridge)
```

```python

print("Linear Regression: ",r2_score(y_test,y_pred_lr))
print("Ridge: ",r2_score(y_test,y_pred_ridge))
print("Lasso: ",r2_score(y_test,y_pred_lasso))
```
